```python
import numpy as np
from scipy import special

x = 1

n_num = np.arange(0, 100)
num = x**(2*n_num + 1)
den = 2*n_num + 1
den = special.factorial(den)

div = num/den

print(div.sum())
```

```python
import numpy as np
import matplotlib.pyplot as plt

n = 100000

x = np.random.rand(n)
y = np.random.rand(n)

xc = np.linspace(0, 1, 200)
yc = (1-xc**2)**(1/2)

dentro = x**2 + y**2 < 1
fora = x**2 + y**2 > 1


plt.plot(xc, yc)
plt.scatter(x[fora], y[fora], color='black')
plt.scatter(x[dentro], y[dentro], color='yellow')
plt.legend('sim')

print(4*dentro.sum()/n)
```

```python
import numpy as np
import matplotlib.pyplot as plt

n = 10000

x = np.random.rand(n)
y = np.random.rand(n)

xr = np.array([0.5] * 200)
yr = np.linspace(0,1,200)

esq = x < 0.5
dir = x > 0.5

print(x[esq].size)
print(x[dir].size)

plt.scatter(x[esq], y[esq])
plt.scatter(x[dir], y[dir])
plt.plot(xr, yr, color='black')
```

```python
import numpy as np
from scipy.special import factorial

x = 270

n = np.arange(1000000)

numerador = ((-1)**n) * (x**(2*n + 1))
denominador = (2*n+1)
denominador = factorial(denominador)

div = numerador / denominador

print(div.sum())
```

```python
import numpy as np
import math
from scipy.special import factorial

x = 10

n = np.arange(1000)

num = x**(2*n)
den = 2*n
den = factorial(den)

cosh = (num/den).sum()

senh = math.sqrt(cosh**2 - 1)

print(senh)
```
